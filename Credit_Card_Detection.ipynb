```python
"""
A simple machine learning project that detects credit card fraud using logistic regression.
 The dataset is balanced by downsampling legitimate transactions, and the model is evaluated based on accuracy on both training and test sets.
 """
```

```python
#Import dependencies
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading the dataset to pd Dataframe
df = pd.read_csv('creditcard.csv')
```

```python
# first 5 rows
df.head()
```

```python
# last five rows
df.tail()
```

```python
# Dataset information
df.info()
```

```python
# check for missing values in each column
df.isnull().sum()
```

```python
# Check distribution of legit transaction and fradulent transactions
df['Class'].value_counts()
```

This dtaset was very unbalanced
0 --> Normal Transaction
1 --> Fraudulent Transaction

```python
# Seprating the data for analysis
legit = df[df.Class == 0]
fraud = df[df.Class == 1]
```

```python
# Printing the shape
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both transactions
df.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing similar distribution of normal and Fraudulent Transactions

Number of Fraudulent transactions --> 492

```python
legit_sample = legit.sample(n=492)
```

Conctenating the two datframes

```python
new_df = pd.concat([legit_sample, fraud], axis=0)
```

```python
# check the first 5 rows of the new df
new_df.head()
```

```python
new_df.tail()
```

```python
new_df['Class'].value_counts()
```

```python
new_df.groupby('Class').mean()
```

```python
# Splitting the data into features and targets
X = new_df.drop(columns='Class', axis=1)
Y = new_df['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
#Splitting the data into training data and testing data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, stratify=Y, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

```python
# Train the model
model = LogisticRegression()
```

```python
# Trining the logistic regression model with Training Data
model.fit(X_train, Y_train)
```

Model evaluation based on accuracy score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
